```python
import numpy as np
import matplotlib.pyplot as plt

def weierstrass(x, a, b, n_terms):
    series = np.zeros_like(x)
    for n in range(n_terms):
        series += a**n * np.cos(np.pi * b**n * x)
    return series

x = np.linspace(-4, 4, 10000)
y = weierstrass(x, a=0.5, b=3, n_terms=30)
plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Weierstrass Function")
plt.xlabel("x")
plt.ylabel("W(x)")
plt.show()
```

```python
num_points = 300
a = 1.5
x_vals = np.linspace(-a, a, num_points)
y_vals = np.linspace(-a, a, num_points)
X, Y = np.meshgrid(x_vals, y_vals)

def f(X, Y):
    return (X ** 2 + Y ** 2 < 1)

v = f(X, Y)

# Plot the heat map
plt.figure(figsize=(6, 6))
plt.imshow(v, extent=[-a, a, -a, a], origin='lower', cmap='gray', aspect='auto')
plt.colorbar(label='f(x, y)')
plt.title('Heatmap of f(x, y) = sin(pi*x) * cos(pi*y)')
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal')
plt.show()
```
